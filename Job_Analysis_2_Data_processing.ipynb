```python
def extract_language(text):
    #print(text)
    #text = text.to_json(orient='records', force_ascii=False)
    ##text=loads(text)
    #text=text[0]
    let_list = []
    for i in range(len(text)):
        if text[i] not in "[{',:}]":
            let_list.append(text[i])
    let_list = let_list[3:]
    let_list = "".join(let_list)
    let_list = let_list.replace('name', '')
    let_list = let_list.replace('level id', '')
    let_list = let_list.split('  ')
    if len(let_list)>=3:
        let_list.pop(1)
        let_list.pop(2)
        return let_list[0]
```

```python
def extract_language_level(text):
    #print(text)
    #text = text.to_json(orient='records', force_ascii=False)
    ##text=loads(text)
    #text=text[0]
    let_list = []
    for i in range(len(text)):
        if text[i] not in "[{',:}]":
            let_list.append(text[i])
    let_list = let_list[3:]
    let_list = "".join(let_list)
    let_list = let_list.replace('name', '')
    let_list = let_list.replace('level id', '')
    let_list = let_list.split('  ')
    if len(let_list)>=3:
        let_list.pop(1)
        let_list.pop(2)
        return let_list[1]
```
